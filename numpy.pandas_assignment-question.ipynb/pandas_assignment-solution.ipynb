### Q22. Calculate the 90% confidence interval of the means for the following:


```python
# Input
import numpy as np
np.random.seed(100)
arr = pd.Series(np.random.normal(10, 3, (100)))
arr.head()
```

__Procedure:__
1. Bootstrap sample a large number of times (10000)
2. Calc mean of each sample.
3. Compute the 5%ile and 95%ile for the lower and upper bounds of the confidence intervals.


````python
# Input
np.random.seed(100)
arr = pd.Series(np.random.normal(10, 3, (100)))
arr.head()
````

````python
# Solution
means = []
for i in range(10000):
    means.append(arr.sample(frac=1, replace=True).mean())

# Sort
means = sorted(means)
means = pd.Series(means)

# lower and upper
means.quantile(q=.05), means.quantile(q=.95) 
````

### Q36. Create an empty dataframe and concatenate the Cars93 dataset for all Manufacturers one by one using a for loop

````python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
````

````python
# Solution
colnames_df = df.columns

new_df = pd.DataFrame(columns = colnames_df)

for manu in df['Manufacturer'].unique():
    temp = df[df['Manufacturer'] == manu]
    temp.reset_index(drop = True, inplace = True)
    new_df = pd.concat([new_df, temp])
    new_df.reset_index(drop = True, inplace = True)
    
new_df
````

### 25. Create a function to compute the coeffecient of variation and apply it row wise in the iris dataset (optional)

`````python
def cv(x):
    return np.std(x) / np.mean(x)


print('Row wise: \n', np.apply_along_axis(cv, 1, arr=iris_2d), '\n')
`````
